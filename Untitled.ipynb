```python
import import_data as imd
from nba_api.stats.static import players, teams
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 6.1; WOW64; rv:64.0) Gecko/20100101 Firefox/64.0"} 
team_list = [] 
for team_info in teams.get_teams(): 
    team_list.append(team_info['id']) 
    
# with respect to defender's proximity 

uri = 'https://stats.nba.com/stats/leaguedashteamstats?Conference=&DateFrom=&DateTo=&Division=&GameScope=&GameSegment=&LastNGames=0&LeagueID=00&Location=&MeasureType=Base&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season=2018-19&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange=&StarterBench=&TeamID=0&VsConference=&VsDivision=' 
tab = imd.get_table_from_url(uri, headers)
```

```python
tab.columns
```

```python
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 6.1; WOW64; rv:64.0) Gecko/20100101 Firefox/64.0"} 
df_list = [] 
for t_id in team_list: 
    sample = imd.get_team_sht_dashboard(4, str(t_id), headers) 
    df_list.append(sample) 
merged_df = pd.concat(df_list) 
pivot_res = pd.pivot_table(merged_df, index='TEAM_NAME', columns = ['CLOSE_DEF_DIST_RANGE'], values = ['FGA_FREQUENCY', 'FG2A_FREQUENCY', 'FG3A_FREQUENCY']) 
fga_freq = pivot_res.loc[:, 'FGA_FREQUENCY'] 
fga_freq = fga_freq.sort_values('6+ Feet - Wide Open', ascending=False)
```

```python
fga_freq = fga_freq.sort_values('6+ Feet - Wide Open', ascending=True)
fga_freq = fga_freq.sort_values('0-2 Feet - Very Tight', ascending=False)
fig = plt.figure(figsize=(15, 10))
plt.title('Shot distribution by defence proximity', color='black', fontsize=20)

fga_freq.plot(kind='barh', stacked=True, ax=fig.gca())
plt.xticks(fontsize=10)
plt.yticks(fontsize=15)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.3), fontsize=15)
plt.show()
```
